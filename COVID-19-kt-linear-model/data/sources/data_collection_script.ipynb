```
import pandas as pd
```

```
jhu_df = pd.read_csv('/content/time_series_19-covid-Confirmed.csv')
jhu_df.head() 
```

```
jhu_df.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace=True)
```

```
jhu_df.head()
```

```
data = pd.DataFrame(columns=jhu_df.columns)
data.head()
```

```
type(jhu_df[jhu_df['Country/Region'] == 'US'].drop(['Country/Region'], axis=1).sum())
```

```
i = 0
for country in jhu_df['Country/Region'].unique():
    if country == 'Taiwan*' or country == 'Cape Verde' or country == 'The Bahamas' \
    or country == 'The Gambia' or country == 'Republic of the Congo' or country == 'East Timor':
        continue
    temp = jhu_df[jhu_df['Country/Region'] == country].drop(['Country/Region'], axis=1).sum()
    data.loc[i] = temp
    data.loc[i, 'Country/Region'] = country
    i += 1
```

```
data.head()
```

```
data.loc[data['Country/Region'] == 'Bahamas, The', 'Country/Region'] = 'Bahamas'
data.loc[data['Country/Region'] == 'Gambia, The', 'Country/Region'] = 'Gambia'
data.loc[data['Country/Region'] == 'Korea, South', 'Country/Region'] = 'South Korea'
```

```
data.drop(['Country/Region'], axis=1).astype('int').to_csv('/content/data_3_23.txt', index=False, header=False)
data['Country/Region'].to_csv('/content/countries_list.txt')

lines = ''
with open('/content/countries_list.txt') as f:
    for line in f:
        country = line.split(',', maxsplit=1)[1].strip().strip('\"').strip('*')
        lines += country + '\n'

with open('/content/countries_list.txt', 'w') as f:
    f.write(lines)
```
